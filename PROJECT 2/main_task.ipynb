Part 6 – Feature Engineering

```python
# Create is_elite column: 1 if overall >= 85, else 0
df['is_elite']=np.where(df['overall'] >= 85, 1, 0)

# Check the first few rows
print(df[['short_name', 'overall', 'is_elite']].head(10))
```

```python
df['value_million']=df['value_eur'] / 1_000_000

# Display the first few rows
print(df[['short_name', 'value_eur', 'value_million']].head(10))
```

```python
df['wage_thousand']=df['wage_eur'] / 1000
print(df[['short_name','wage_eur','wage_thousand']].head(10))
```

```python
# Step 1: Calculate growth
df['growth']=df['potential'] - df['overall']

# Step 2: Create high_growth column
df['high_growth']=np.where(df['growth']>5,1,0)

# Display first 10 rows
print(df[['short_name','overall','potential','growth','high_growth']].head(10))
```

```python
# Correct way to extract primary position
df['primary_positions']=df['player_positions'].str.split(',').str[0]

# Display first 10 rows
print(df[['short_name', 'primary_positions']].head(10))
```

Predict Young Players Likely to Become Elite

```python
# Filter players under 23 with high potential
young_promising=df[(df['age'] < 23) & (df['potential'] >= 85)]

# Sort by potential and growth to find top prospects
top_young_players=young_promising.sort_values(by=['potential','age'], ascending=False)

# Display top 10 young players
print(top_young_players[['short_name','age','overall','potential']].head(10))
```

```python
# Create Experience_Score
df['Experience_Score']=df['overall'] - abs(22 - df['age'])

# Display top 10 players by Experience_Score
top_experienced = df.sort_values(by='Experience_Score', ascending=False)
print(top_experienced[['short_name', 'age', 'overall', 'Experience_Score']].head(10))
```
